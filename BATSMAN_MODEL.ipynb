**Import Libraries**

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

**Import Dataset**

```
dataset = pd.read_csv('BATSMAN_STATS.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

**Split data to training and testing**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state= 0)
```

**Fit the data to regression model** 

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

**Predict and print the data**

```
y_pred = lr.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

**Add constants to X dataset for OLS** 

```
import statsmodels.api as sm
X = np.append(arr = np.ones((61, 1)).astype(int), values = X, axis = 1)
print(X)
```

**Do backward Elimination for identifying best fit data**

```
X_opt = X[:, [0, 1, 2, 3, 4]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```
X_opt = X[:, [0,  2, 3, 4]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

**Print Predicted Values**

```
y_pred_ols = regressor_OLS.predict()
np.set_printoptions(precision=2)
print(np.concatenate((y_pred_ols.reshape(len(y_pred_ols),1), y.reshape(len(y),1)),1))
```
