**Import Libararies**


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Import data to the process**

```
dataset = pd.read_csv('CORONA_TN_DATA.csv')
D = dataset.iloc[:, : 1].values
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

**Train and Test split the data**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
D_train, D_test = train_test_split(D, test_size=0.3, random_state = 1)
```

**Implement the Linear Regression Model for Test case and Positive Ratio**

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

**Implement the Linear Regression Model for Days and Number of Test Cases**

```
from sklearn.linear_model import LinearRegression
dr = LinearRegression()
dr.fit(D_train, X_train)
```

**Predict and Plot with Training set data**

```
Y_pred =  lr.predict(X_test)
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train,lr.predict(X_train), color='blue')
plt.title('Corona Tests vs Positive data(Training set)')
plt.xlabel('Number of tests')
plt.ylabel('Number of Positives')
plt.show()

```
